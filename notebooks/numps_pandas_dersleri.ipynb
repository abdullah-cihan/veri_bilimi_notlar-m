```python
import numpy as np

notlar = np.array([[50, 60],  # 1. Öğrencinin Vize ve Finali
                   [70, 80],  # 2. Öğrencinin Vize ve Finali
                   [90, 100]]) # 3. Öğrencinin Vize ve Finali

# Tüm sınıftaki tüm notların genel ortalaması (Eksen belirtmezsek hepsini tek bir dizi gibi düşünür)
genel_ortalama = notlar.mean() 

sinav_ortalamalari = notlar.mean(axis=0)
sinav_ortalamalari

ogrenci_bazli=notlar.mean(axis=1,keepdims=True)
ogrenci_bazli
```

```python
satislar = np.array([[100, 50, 20],
                     [120, 60, 30],
                     [ 90, 40, 10]])

satis_top=satislar.sum().copy()
print(satis_top)

gun_bazli=satislar.sum(axis=0,keepdims=True)
print(gun_bazli)
```
